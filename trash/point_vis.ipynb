```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.view_init(elev=25, azim=0)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)

ax.scatter(0, 0, 0)

with open("cpp_output.txt") as f:
    for line in f.readlines():
        x, y, z = [float(val) for val in line[1:-2].split(", ")]
        ax.scatter(x, y, z)
```
